```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
```

```python
x = np.arange(70, 191, 2)
y = np.zeros_like(x)

y[x >120] = 1
print(x.shape)
print(y.shape)
```

```python
print("BP: \n", x)
```

```python
print("High BP?\n", y)
```

```python
from matplotlib.colors import ListedColormap
colors = ["blue", "red"]
cmap = ListedColormap(colors[:len(np.unique(y))])
```

```python
f=plt.figure(figsize=(20,10))
ax = f.add_subplot(111)
ax.scatter(x, y, c=y, cmap=cmap)
ax.legend()
ax.set_ylabel("Target")
ax.set_xlabel("Systolic Blood Pressure")
f.tight_layout()
plt.show()
```

```python
def sigmoid(v):
    return 1/(1+np.exp(-v))

v = np.arange(-10, 10, .2)
sig_v = sigmoid(v)    

f = plt.figure(figsize=(20, 10))
plt.plot(v, sig_v)
plt.xlabel("V",)
plt.ylabel("Sigmoid")
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(20, 10))

def plot_sigmoid(x, ms, cs, n):
    
    for i in range(len(ms)):
        ax = fig.add_subplot(int(np.ceil(np.sqrt(n))), int(np.round(np.sqrt(n))), i+1)
        ax.plot(x, sigmoid( ms[i]*x + cs[i]), linewidth=2)
        ax.set_xlabel("$Input$")
        ax.set_ylabel("Sigmoid")
        ax.set_title("$m=%0.4s, c=%0.6s$"%(ms[i], cs[i]))
    fig.tight_layout()
    plt.show()

x = np.arange(-10, 10, 0.1)
ms = [10, 5, 2, 1, 1/3, 1/6, 1/12, 1/15, 1/18] 
cs = [-1210, 5*-121, -242, -121, -121/3, -121/6, -121/12, -121/15, -121/18]

plot_sigmoid(x, ms, cs, len(ms))
```
